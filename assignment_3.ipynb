# 3️⃣ Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums):
        freq = {}
        count = 0

        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1

        return count
```

# 4️⃣ How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        result = []

        for i in nums:
            count = 0
            for j in nums:
                if j < i:
                    count += 1
            result.append(count)

        return result
```

# 5️⃣ Subtract the Product and Sum of Digits

```python
class Solution:
    def subtractProductAndSum(self, n):
        product = 1
        total = 0

        for digit in str(n):
            d = int(digit)
            product *= d
            total += d

        return product - total
```

# 6️⃣ XOR Operation in an Array

```python
class Solution:
    def xorOperation(self, n, start):
        result = 0

        for i in range(n):
            result ^= (start + 2 * i)

        return result
```
