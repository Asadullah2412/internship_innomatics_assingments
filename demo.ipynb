```python
my_str = 'Peter Piper Picked A Peck Of Pickled Peppers.'
sub_str = 'Peck'
new_str = 'Pack'

output = ''
i = 0

while i < len(my_str):
    if my_str[i:i+len(sub_str)] == sub_str:
        output += new_str
        i += len(sub_str)
    else:
        output += my_str[i]
        i += 1

print(output)
```

```python
print('bubble sort')
l1 = [100,10,1,298,65,483,49876,2,80,9,9213]

for i in range(len(l1)):
    for j in range(len(l1)-i-1):
        if l1[j]>l1[j+1]:
            l1[j], l1[j+1] = l1[j+1],l1[j]

print(l1)
```
