# 1️⃣ Maximum Product of Two Elements in an Array

```python

def maxProduct(self, nums):
    nums.sort()
    return (nums[-1] - 1) * (nums[-2] - 1)
```

# 2️⃣ Count Number of Teams

```python

def numTeams(self, rating):
    n = len(rating)
    count = 0

    for j in range(1, n - 1):
        left_smaller = left_larger = 0
        right_smaller = right_larger = 0

        for i in range(j):
            if rating[i] < rating[j]:
                left_smaller += 1
            else:
                left_larger += 1

        for k in range(j + 1, n):
            if rating[k] > rating[j]:
                right_larger += 1
            else:
                right_smaller += 1

        count += left_smaller * right_larger
        count += left_larger * right_smaller

    return count
```

# 3️⃣ Number of Students Doing Homework at a Given Time

```python

def busyStudent(self, startTime, endTime, queryTime):
    count = 0

    for i in range(len(startTime)):
        if startTime[i] <= queryTime <= endTime[i]:
            count += 1

    return count
```

# 4️⃣ Number of Steps to Reduce a Number to Zero

```python

def numberOfSteps(self, num):
    steps = 0

    while num > 0:
        if num % 2 == 0:
            num //= 2
        else:
            num -= 1
        steps += 1

    return steps
```

# 5️⃣ Counting Bits

```python

def countBits(self, n):
    bits = [0] * (n + 1)

    for i in range(1, n + 1):
        bits[i] = bits[i // 2] + (i % 2)

    return bits
```
