```python
def wornw(num):# function name is weird or not weird 
    if num %2 ==0:
        if num>=2 and num <5:
            print('Not Weird')
        elif num >=6 and num <20:
            print('Weird')
        elif num >20:
            print('Not Weird')
    elif num %2 != 0:
        print('Weird')

wornw(24)
```
